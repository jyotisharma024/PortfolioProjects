# **Check for the Missing Values**

```python
# Generally in a Numerical Continous column we fill the missing values with the mean or median
# In categorical column we fill with Mode
```

```python
data.isna().sum()
```

```python
data.nunique()
```

```python
num_col=[]
cat_col=[]
for i in data.columns:
    if data[i].nunique()>5 and data[i].dtype==float or data[i].dtype==int:
        num_col.append(i)
    else:
        cat_col.append(i)
```

```python
num_col
```

```python
cat_col
```

```python
def null_impute_numeric(column):
    data[column].fillna(data[column].mean(),inplace=True)
```

```python
def null_impute_categoric(column):
    data[column].fillna(data[column].mode()[0],inplace=True)
```

```python
for i in num_col:
    null_impute_numeric(i)
```

```python
for i in cat_col:
    null_impute_categoric(i)
    
```

```python
data.isna().sum()
```

 # **Treating the Outliers**

*Since the last two columns in num_col are categorical columns so we have tom remove them from the num_col list.*

```python
num_col.pop()
```

```python
num_col
```

```python
def cap_outliers(col):
    q3=data[col].quantile(.75)
    q1=data[col].quantile(.25)

    iqr=(q3-q1)

    upper=q3+1.5*iqr
    lower=q3-1.5*iqr

    data[col].clip(lower,upper, inplace=True)

```

```python
for i in num_col:
    cap_outliers(i) 
```

```python
f,ax=plt.subplots(2,4, figsize=(22,8),dpi=100)
ind=0
for i in range(2):
    for j in range(4):
        sns.scatterplot(y=data[num_col[ind]],x=data[num_col[ind]], ax=ax[i,j])
        ind+=1
    plt.tight_layout()   
```
